```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Sample financial statement data (offline)
data = {
    'Year': [2020, 2021, 2022],
    'Revenue': [1000000, 1200000, 1400000],
    'COGS': [400000, 500000, 600000],
    'Operating_Expenses': [300000, 320000, 350000],
    'Net_Income': [180000, 220000, 260000],
    'Total_Assets': [800000, 900000, 1000000],
    'Shareholder_Equity': [400000, 450000, 500000],
    'Current_Assets': [300000, 350000, 400000],
    'Current_Liabilities': [150000, 170000, 200000]
}

# Step 3: Create DataFrame
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

# Step 4: Calculate ratios
df['Gross_Profit'] = df['Revenue'] - df['COGS']
df['Operating_Income'] = df['Gross_Profit'] - df['Operating_Expenses']

df['Net_Profit_Margin'] = df['Net_Income'] / df['Revenue']
df['Gross_Profit_Margin'] = df['Gross_Profit'] / df['Revenue']
df['Operating_Margin'] = df['Operating_Income'] / df['Revenue']
df['ROE'] = df['Net_Income'] / df['Shareholder_Equity']
df['Current_Ratio'] = df['Current_Assets'] / df['Current_Liabilities']

# Step 5: Show summary ratios
ratios = df[['Net_Profit_Margin', 'Gross_Profit_Margin', 'Operating_Margin', 'ROE', 'Current_Ratio']]
print(ratios)

# Step 6: Plot Net Profit Margin over time
ratios['Net_Profit_Margin'].plot(kind='line', marker='o', title='Net Profit Margin Over Time')
plt.ylabel('Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()
```
